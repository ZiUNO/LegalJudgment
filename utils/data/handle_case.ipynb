```python
# 读取处理后案例文件生成数据集
train_dataset_rate = 7/10  # 修改训练集比例调节
train_json = {}
dev_json = {}

handled_dir_path = os.path.join(case_path, 'handled')
label_data = []

for handle_file_name in os.listdir(handled_dir_path):
    tsv_data = []
    with open(os.path.join(handled_dir_path, handle_file_name), 'r', encoding='utf-8') as f:
        data = json.load(f)
    index = str(len(label_data))
    label_data.append(handle_file_name.replace("handled_", "").replace(".json", ""))
    for uniqid in data:
        tsv_data.append({"uniqid": uniqid, "strContent": [content["strContent"] for content in data[uniqid]]})
    shuffle(tsv_data)
    train_dataset_size = int(train_dataset_rate * len(tsv_data))
    def form_dataset_dict(dataset, tsv, index):
        for data in tsv:
            if data["uniqid"] not in list(dataset.keys()):
                dataset[data["uniqid"]] = {"strContent": data["strContent"], "kind": {index}}
            else: 
                dataset[data["uniqid"]]["kind"].add(index)
        return dataset
    train_json = form_dataset_dict(train_json, tsv_data[:train_dataset_size], index)
    dev_json = form_dataset_dict(dev_json, tsv_data[train_dataset_size:], index)

def json2dataset(json_data):
    dataset = []
    for uniqid in json_data:
        for content in json_data[uniqid]["strContent"]:
            dataset.append([uniqid, content, str(list(json_data[uniqid]['kind']))])
    shuffle(dataset)
    return dataset
    
def dataset2tsv(dataset):
    tsv_data = [["uniqid", "strContent", "kind"]] + dataset
    tsv_data = ["\t".join(line) + "\n" for line in tsv_data]
    tsv_data[-1] = tsv_data[-1].strip()
    return tsv_data
    
    
def save_tsv(tsv, file_name):
    with open(file_name, 'w', encoding="utf-8", newline="\n") as f:
        f.writelines(tsv)
    
    
jsons = {"train": train_json, "val": dev_json}
dataset = {key: json2dataset(jsons[key]) for key in jsons}
tsv = {key: dataset2tsv(dataset[key]) for key in dataset}
multi_dir_name = 'multi_data_dir'
multi_dir_path = os.path.join(case_path, 'multi_data_dir')
if not Path(multi_dir_path).exists():
    os.makedirs(multi_dir_path)
_ = [save_tsv(tsv[key], os.path.join(multi_dir_path, "%s.tsv" % key)) for key in tsv]
```
